```python
def get_highlight_column(col_df):
    col = list()
    for i in col_df:
        if i == "{}":
            col.append(set())
        else:
            col.append(set([int(j) for j in i.split(",")]))
    return col
```

```python
Sentence1_Highlighted_1 = get_highlight_column(test_df["Sentence1_Highlighted_1"])
Sentence2_Highlighted_1 = get_highlight_column(test_df["Sentence2_Highlighted_1"])
Sentence1_Highlighted_2 = get_highlight_column(test_df["Sentence1_Highlighted_2"])
Sentence2_Highlighted_2 = get_highlight_column(test_df["Sentence2_Highlighted_2"])
Sentence1_Highlighted_3 = get_highlight_column(test_df["Sentence1_Highlighted_3"])
Sentence2_Highlighted_3 = get_highlight_column(test_df["Sentence2_Highlighted_3"])
```

```python
len_data = len(Sentence1_Highlighted_1)
index_good = list()
threshold = 0.5

for i in range(len_data):
    total_len_sentence1 = len(Sentence1_Highlighted_1[i]) + len(Sentence1_Highlighted_2[i]) + len(Sentence1_Highlighted_3[i])
    total_len_sentence2 = len(Sentence2_Highlighted_1[i]) + len(Sentence2_Highlighted_2[i]) + len(Sentence2_Highlighted_3[i])
    total_len_sentence = max(1,total_len_sentence1 + total_len_sentence2)
    inter_sentence1 = Sentence1_Highlighted_1[i].intersection(Sentence1_Highlighted_2[i],Sentence1_Highlighted_3[i])
    inter_sentence2 = Sentence2_Highlighted_1[i].intersection(Sentence2_Highlighted_2[i],Sentence2_Highlighted_3[i])
    if  len(inter_sentence1) + len(inter_sentence2)/ total_len_sentence > threshold:
        index_good.append(i)

print(len(index_good) / len_data)
```
